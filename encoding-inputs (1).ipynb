```python
import pandas as pd
import numpy as np
import os
from tensorflow.keras.preprocessing.text import Tokenizer
import pickle
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# For reading previous data
csv_data = pd.read_csv('/kaggle/input/nlp-project-v2/Grammer_sentences_v2.csv')

x = csv_data['Input'].to_numpy()

x_l = list(x)
```

```python
# loading tokenizer
with open('/kaggle/input/encoding-1/grammar_tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
```

```python
x_text_vector = tokenizer.texts_to_sequences(x_l)
```

```python
x_text_vector[2]
```

```python
tokenizer.sequences_to_texts([x_text_vector[2]])
```

**Padding input sentences to a unit size (50 word in each sentence)**

```python
max_length = 50
x_padded = pad_sequences(x_text_vector, maxlen=max_length,)
```

```python
x_padded.shape
```

```python
tokenizer.sequences_to_texts([x_padded[7]])[0].strip('<OOV>').strip(' <OOV> ')
```

```python
# Save input vector in csv file 
np.savetxt("Input_vector.csv", x_padded, fmt="%d", delimiter=",")
```
